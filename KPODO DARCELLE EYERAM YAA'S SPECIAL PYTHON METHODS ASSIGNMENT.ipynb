# Python Special Methods

Special methods (also called dunder methods) have double underscores like `__init__`.
They let your custom classes work with Python's built-in features.

## 1. `__init__` — Create an Object

Runs automatically when you make a new object. Sets up starting values.

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

buddy = Dog('Buddy', 3)
print(buddy.name, buddy.age)
```

## 2. `__str__` — Print Nicely

Controls what `print()` shows. For users.

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def __str__(self):
        return f"{self.name} is {self.age} years old"

buddy = Dog('Buddy', 3)
print(buddy)
```

## 3. `__repr__` — Debug View

Shows info for developers. Should help recreate the object.

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def __repr__(self):
        return f"Dog('{self.name}', {self.age})"

buddy = Dog('Buddy', 3)
print(repr(buddy))
```

## 4. `__len__` — Get Length

Lets `len()` work on your object.

```python
class Team:
    def __init__(self, members):
        self.members = members
    
    def __len__(self):
        return len(self.members)

team = Team(['Alice', 'Bob', 'Carol'])
print(len(team))
```

## 5. `__eq__` — Check Equality

Defines what `==` means for your objects.

```python
class Box:
    def __init__(self, color):
        self.color = color
    
    def __eq__(self, other):
        return self.color == other.color

b1 = Box('red')
b2 = Box('red')
b3 = Box('blue')

print(b1 == b2)
print(b1 == b3)
```

## 6. `__lt__` — Less Than

Lets you sort objects with `<`.

```python
class Score:
    def __init__(self, value):
        self.value = value
    
    def __lt__(self, other):
        return self.value < other.value
    
    def __repr__(self):
        return str(self.value)

scores = [Score(50), Score(20), Score(80)]
scores.sort()
print(scores)
```

## 7. `__add__` — Add Objects

Teaches `+` how to work with your class.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)
    
    def __str__(self):
        return f"({self.x}, {self.y})"

p1 = Point(1, 2)
p2 = Point(3, 4)
print(p1 + p2)
```

## 8. `__getitem__` — Get by Index

Lets you use `[]` to access items.

```python
class Team:
    def __init__(self, members):
        self.members = members
    
    def __getitem__(self, index):
        return self.members[index]

team = Team(['Alice', 'Bob', 'Carol'])
print(team[0])
print(team[2])
```

## 9. `__setitem__` — Set by Index

Lets you assign with `[]`.

```python
class Team:
    def __init__(self, members):
        self.members = members
    
    def __setitem__(self, index, value):
        self.members[index] = value
    
    def __str__(self):
        return str(self.members)

team = Team(['Alice', 'Bob', 'Carol'])
team[1] = 'Ben'
print(team)
```

## 10. `__contains__` — Check Membership

Lets you use `in` with your object.

```python
class Team:
    def __init__(self, members):
        self.members = members
    
    def __contains__(self, name):
        return name in self.members

team = Team(['Alice', 'Bob', 'Carol'])
print('Bob' in team)
print('Dave' in team)
```

## 11. `__call__` — Call Like a Function

Makes your object callable with `()`.

```python
class Greeter:
    def __init__(self, greeting):
        self.greeting = greeting
    
    def __call__(self, name):
        return f"{self.greeting}, {name}!"

hi = Greeter('Hi')
print(hi('Alice'))
print(hi('Bob'))
```

## 12. `__iter__` & `__next__` — Loop Over Objects

Lets you use `for` loops on your object.

```python
class Countdown:
    def __init__(self, start):
        self.start = start
    
    def __iter__(self):
        self.current = self.start
        return self
    
    def __next__(self):
        if self.current < 0:
            raise StopIteration
        num = self.current
        self.current -= 1
        return num

for n in Countdown(5):
    print(n)
```

## 13. `__bool__` — True or False

Controls what happens in `if` statements.

```python
class Wallet:
    def __init__(self, money):
        self.money = money
    
    def __bool__(self):
        return self.money > 0

empty = Wallet(0)
full = Wallet(50)

print(bool(empty))
print(bool(full))
```

## 14. `__del__` — Cleanup

Runs when an object is deleted. Don't rely on it for important cleanup.

```python
class Temp:
    def __init__(self, name):
        self.name = name
        print(f"Created: {name}")
    
    def __del__(self):
        print(f"Deleted: {self.name}")

t = Temp('test')
del t
```

## 15. `__enter__` & `__exit__` — With Statement

Makes your object work with `with` for safe resource handling.

```python
class Timer:
    def __enter__(self):
        print('Timer started')
        return self
    
    def __exit__(self, a, b, c):
        print('Timer stopped')

with Timer():
    print('Doing work...')
```

## 16. `__hash__` — Use in Sets and Dicts

Lets your object be a key in dictionaries or in sets.

```python
class Person:
    def __init__(self, id, name):
        self.id = id
        self.name = name
    
    def __eq__(self, other):
        return self.id == other.id
    
    def __hash__(self):
        return hash(self.id)
    
    def __repr__(self):
        return self.name

p1 = Person(1, 'Alice')
p2 = Person(1, 'Alice')
p3 = Person(2, 'Bob')

people = {p1, p2, p3}
print(people)
```

## Quick Reference

| Method | Used For | Example |
|---|---|---|
| `__init__` | Make object | `Dog('Buddy', 3)` |
| `__str__` | Print nicely | `print(dog)` |
| `__repr__` | Debug info | `repr(dog)` |
| `__len__` | Count items | `len(team)` |
| `__eq__` | Compare equal | `a == b` |
| `__lt__` | Compare less | `a < b` |
| `__add__` | Add objects | `a + b` |
| `__getitem__` | Get item | `obj[0]` |
| `__setitem__` | Set item | `obj[0] = x` |
| `__contains__` | Check inside | `x in obj` |
| `__call__` | Call object | `obj()` |
| `__iter__`/`__next__` | Loop | `for x in obj` |
| `__bool__` | True/False | `if obj:` |
| `__del__` | Cleanup | `del obj` |
| `__enter__`/`__exit__` | Safe block | `with obj:` |
| `__hash__` | Dict key | `hash(obj)` |
